#Step6: Design k-NN classifier using hyper-parameter tuning and model selection

```python
import scipy.spatial.distance as distance
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
# user defined chi_square distance
def chi_square_distance(X, Y):
    with np.errstate(divide="ignore", invalid="ignore"):
        return (1 / 2) * np.sum(np.nan_to_num((np.square(X - Y) / (X + Y))))
```

```python
n_neighbors = [int(x) for x in np.linspace(1, 15, num = 15)]
metric = [chi_square_distance, distance.euclidean, distance.minkowski, distance.cosine]
param_distribs = {'n_neighbors':n_neighbors,'metric': metric}
estimator = KNeighborsClassifier()
```

```python
## we are using repeated cross-validation here. This means the cross-validation will be repeatedly run multiple times to get more robust error estimation. The key changes are highlighted
from sklearn.model_selection import RepeatedKFold
skfold = RepeatedKFold(n_splits=10,n_repeats=10,random_state=1)
grid_search_blood = GridSearchCV(estimator, param_distribs, n_jobs=-1,cv=skfold)
grid_search_cancer = GridSearchCV(estimator, param_distribs, n_jobs=-1,cv=skfold)
grid_search_ecoli = GridSearchCV(estimator, param_distribs, n_jobs=-1,cv=skfold)
```

```python
# Grid search for blood
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_blood = blood_df.drop('whether he/she donated blood in March 2007', axis=1)
X_train_blood = scaler.fit_transform(X_train_blood)
y_train_blood = blood_df['whether he/she donated blood in March 2007']
grid_search_blood.fit(X_train_blood, y_train_blood)
```

```python
# Grid search for breast cancer
scaler = MinMaxScaler()
X_train_cancer = scaler.fit_transform(cancer_features_encode)
y_train_cancer = cancer_label
grid_search_cancer.fit(X_train_cancer, y_train_cancer)
```

```python
# Grid search for ecoli
scaler = MinMaxScaler()
X_train_ecoli = ecoli_df_update.drop('Label', axis=1)
X_train_ecoli = scaler.fit_transform(X_train_ecoli)
y_train_ecoli = ecoli_df_update['Label']
grid_search_ecoli.fit(X_train_ecoli, y_train_ecoli)
```

String processing for Metric

```python
import re
```

```python
# Function for creating a new tabel with metric feature
def create_pd_with_metric(df):
  df_copy = df.copy()
  df = df['param_metric'].to_string()
  df = re.sub(r'\d+', '', df)
  df = re.sub(r'\n', '', df)
  df = df.strip()
  df = df.split("<function")
  df = df[1:]

  # Create a new table for blood
  mertic = []
  metric.clear()
  for string in df:
    metric.append(string.split('at')[0].strip())

  data = {'param_n_neighbors':df_copy['param_n_neighbors'], 'param_metric':df_copy['param_metric'], 'mean_test_score':df_copy['mean_test_score'], 'metric':metric}
  df = pd.DataFrame(data=data)
  df_copy.head()
  
  return df
```

```python
blood_df_metric = create_pd_with_metric(df_cv_blood)
blood_df_metric.head(5)
```

```python
cancer_df_metric = create_pd_with_metric(df_cv_cancer)
```

```python
# Function for Classification accuracy of K-NN over Blood dataset
def figure_plot(dframe, title):
  for label, df in dframe.groupby('metric'):
    plt.plot(df.param_n_neighbors, df.mean_test_score, label=label)
    plt.title(title)
    plt.legend()
```

### 3. Report the best cross-validation score on three datasets. For each dataset, you need to select the best k and distance function, then you can run 'cross_val_score()' function to calculate final cross-validation accuracy.

```python
from sklearn.model_selection import cross_val_score
```

Select best k and distance function

```python
# Function for calculate final cross-validation score
def val_best_model(df, X_train, Y_train):
  best_model = df.best_estimator_
  metric_choosen, n_choosen = best_model.get_params()['metric'], best_model.get_params()['n_neighbors']
  best_model.fit(X_train, Y_train)
  print("Distance choosen:", metric_choosen)
  print("N choosen:", n_choosen)
  print("Accuracy:", cross_val_score(best_model, X_train, Y_train, cv=10))
```

```python
# Calculate final cross-validation score for blood dataset
val_best_model(grid_search_blood, X_train_blood, Y_train_blood)
```

```python
# Calculate final cross-validation score for cancer dataset
val_best_model(grid_search_cancer, X_train_cancer, Y_train_cancer)
```

```python
# Calculate final cross-validation score for ecoli dataset
val_best_model(grid_search_ecoli, X_train_ecoli, Y_train_ecoli)
```

#Step8: You need to average classification accuracy of k-NN over three datasets and visualize the results as shown in Figure 8 in the paper.  

Group three data set together by distance metric

```python
def distance_group(distance):
  df = pd.concat([blood_df_metric[blood_df_metric['metric'] == distance], cancer_df_metric[cancer_df_metric['metric'] == distance], ecoli_df_metric[ecoli_df_metric['metric'] == distance]])
  return df
```

```python
chi_square_df = distance_group('chi_square_distance')
```

```python
euclidean_df = distance_group('euclidean')
```

```python
cosine_df = distance_group('cosine')
```

```python
minkowski_df = distance_group('minkowski')
```

```python
chi_square_df.groupby(chi_square_df['param_n_neighbors']==1).mean()['mean_test_score'][True]
```

```python
chi_square_mean = chi_square_df.groupby('param_n_neighbors').mean()
euclidean_mean = euclidean_df.groupby('param_n_neighbors').mean()
cosine_mean = cosine_df.groupby('param_n_neighbors').mean()
minkowski_mean = minkowski_df.groupby('param_n_neighbors').mean()
```

```python
# Plot the everage of classification accuracy of k-NN over three datasets 
plt.plot(chi_square_mean.index, chi_square_mean.mean_test_score, label='chi-square')
plt.plot(euclidean_mean.index, euclidean_mean.mean_test_score, label='euclidean')
plt.plot(cosine_mean.index, cosine_mean.mean_test_score, label='cosine')
plt.plot(minkowski_mean.index, minkowski_mean.mean_test_score, label='minkowski')
plt.title('The everage of classification accuracy of k-NN over three datasets')
plt.legend()
```
